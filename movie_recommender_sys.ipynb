```python
# we are getting improper list from the genre so we need to process it in such a way that we can get all the name of genres in one list
import ast  # No longer needed, but kept for clarity

def convert(obj):
    
    """Extracts genre names from lists or individual strings."""
    if isinstance(obj, list):  # Check if it's already a list
        return obj
    else:
        L=[] 
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
        
#         return ast.literal_eval(obj)  # Handle cases where it's a string

movies['genres'] = movies['genres'].apply(convert)
```

```python
def convertk(obj):
    """Extracts genre names from nested lists or individual strings."""
    if isinstance(obj, list):
        # Recursively process each element in the list
        l_key = []
        for item in obj:
            if isinstance(item, dict):
                l_key.append(item['name'])
            else:
                # Handle nested lists if needed (example here)
                sub_genres = convert(item)
                if sub_genres:
                    l_key.extend(sub_genres)
        return l_key
    else:
        # Try converting a single string to a dictionary (if applicable)
        try:
            dict_obj = ast.literal_eval(obj)
            return [dict_obj['name']]
        except (ValueError, SyntaxError):
            return []  # Return empty list if conversion fails

# Apply the function
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# we are getting improper list from the genre so we need to process it in such a way that we can get all the name of genres in one list
import ast  # No longer needed, but kept for clarity

def convert3(obj):
    
    """Extracts genre names from lists or individual strings."""
    if isinstance(obj, list):  # Check if it's already a list
        return obj
    else:
        L=[] 
        counter=0
        for i in ast.literal_eval(obj):
            if counter !=3:
                L.append(i['name'])
                counter+=1
            else:
                break
        return L
        
#         return ast.literal_eval(obj)  # Handle cases where it's a string
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# to fetch the director name
def fetch_director(obj):
    
    """Extracts genre names from lists or individual strings."""
    if isinstance(obj, list):  # Check if it's already a list
        return obj
    else:
        L=[] 
        for i in ast.literal_eval(obj):
            if i['job']=='Director':
                L.append(i['name'])
                break

        return L
        
#         return ast.literal_eval(obj)  # Handle cases where it's a string
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# to convert the overview which is in string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# need to apply a tanformation to remove the space in between and use the camelcase
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
# movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])

movies.head()
```

```python
# need to create tag to concatenate overview, genres, keywords, cast, crew
# movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']

# Convert lists to strings before concatenating to get rid of multiple commas and others
def list_to_str(lst):
    if isinstance(lst, list):
        return ' '.join(lst)
    return str(lst)

# it was used because after concatenating there we used to get ( ,, and '''''')
movies['tags'] = (
    movies['overview'].apply(list_to_str) +
    movies['genres'].apply(list_to_str) +
    movies['keywords'].apply(list_to_str) +
    movies['cast'].apply(list_to_str) +
    movies['crew'].apply(list_to_str)
)

# Display the 'tags' column for the first row
print(movies['tags'][0])
```

```python
# need to vectorise the tag as we some words which is common or have good frequency and based on that we will recommend the movie
 #as we dont need stop words like is are am , in, the and the mxt word count we will need is 5000 to have better accuracy
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors= cv.fit_transform(movies['tags']).toarray()
# print(vectors[0])
```

```python
# to get the name or data of the corpus
feature_names = cv.get_feature_names_out()
print(feature_names)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```

```python
stem('Captain Barbossa, long believed to be dead, has come back to life and is headed to the edge of the Earth with Will Turner and Elizabeth Swann. But nothing is quite as it seems.Adventure Fantasy Actionocean drugabuse exoticisland eastindiatradingcompany loveofone slife traitor shipwreck strongwoman ship alliance calypso afterlife fighter pirate swashbuckler aftercreditsstingerJohnnyDepp OrlandoBloom KeiraKnightleyGoreVerbinski')
```

```python
movies['tags']= movies['tags'].apply(stem)
print(movies['tags'])
```

```python
#  we need to calculate distance between each movie, more the distance less the similarity
#  we will not cal euclidian dist instead we cal angle between them(cosine dis) less the angle more the similarity
#  euclidian distance is  not reliable for highre dimension

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# we need to give the most five similar movie
# for this we can take the movie as an input and will go into the matric and will fetch the index of that movie
# after that we can take the five most similar movie with higher similarity can be taken

def recommend_movie(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6] #it will give the tuple of required movies relation with other movies as movie_name and similarity
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend_movie('Spectre')
```
